```python
# Extract homepage hero background image data URI from index.html
from pathlib import Path

html = Path('index.html').read_text(encoding='utf-8')
needle = "background-image: url('"
start = html.find(needle)
if start < 0:
    raise SystemExit('needle not found')
start += len(needle)
end = html.find("')", start)
if end < 0:
    raise SystemExit('end not found')
uri = html[start:end]
print(uri[:200])
Path('hero-bg-uri.txt').write_text(uri, encoding='utf-8')
print('saved hero-bg-uri.txt')
```
